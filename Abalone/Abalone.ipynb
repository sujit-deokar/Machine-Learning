```python
import pandas as pd
feature_names = ['Sex','Length','Diameter','Height','Whole Wt','Shucked Wt','Viscera Wt','Shell Wt','Rings']
```

```python
abalone = pd.read_csv('abalone.data',names = feature_names)
```

```python
abalone.head()
```

```python
abalone.describe()
```

```python
abalone.replace(('M', 'F','I'), (0,1,2), inplace=True)
```

```python
import keras
all_features = abalone[feature_names].drop('Sex',axis=1).values
all_classes = abalone['Sex'].values

all_classes = keras.utils.to_categorical(all_classes,3)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
all_features1 = scalar.fit_transform(all_features)
```

```python
from keras.optimizers import adam
from keras.layers import Dense,Dropout,Embedding
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
import numpy as np
from keras.callbacks import ModelCheckpoint
```

```python
from sklearn.model_selection import train_test_split
from keras.preprocessing import sequence
np.random.seed(1234)

(training_inputs,
 testing_inputs,
 training_classes,
 testing_classes) = train_test_split(all_features1, all_classes, train_size=0.75, random_state=1)
seq_len = 60
testing_inputs_lstm=sequence.pad_sequences(testing_inputs, maxlen=seq_len, padding='post', dtype='float', truncating='post')
training_inputs_lstm=sequence.pad_sequences(training_inputs, maxlen=seq_len, padding='post', dtype='float', truncating='post')
```

```python
model = Sequential()
model.add(Dense(256,activation='relu',input_shape=(8,),kernel_initializer='normal'))
model.add(Dropout(0.4))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(1024,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(32,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(64,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(1024,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(256,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(3,activation='softmax'))
chk = ModelCheckpoint('best_model.pkl', monitor='val_acc', save_best_only=True, mode='max', verbose=1)
model.compile(loss='categorical_crossentropy',
              optimizer = adam(),
              metrics = ['accuracy'])
history = model.fit(training_inputs,training_classes,
                    verbose=1,
                    epochs=50,
                    callbacks=[chk],
                    batch_size = 30,
                    validation_data = (testing_inputs,testing_classes))
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(128,activation='relu',input_shape=(8,)))
    model.add(Dropout(0.4))
    model.add(Dense(256,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(512,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(1024,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(2048,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(1024,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(512,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(256,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(128,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(3,activation='softmax'))
    model.compile(loss='categorical_crossentropy',
                  optimizer = adam(),
                  metrics = ['accuracy'])
    return model
estimator = KerasClassifier(build_fn=create_model,nb_epoch=50,verbose=1)
score = cross_val_score(estimator,all_features1,all_classes,cv=50)
```
