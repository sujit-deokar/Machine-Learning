```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
feature_names = ['Sex','Length','Diameter','Height','Whole Wt','Shucked Wt','Viscera Wt','Shell Wt','Rings']
data = pd.read_csv('abalone.data',names=feature_names,na_values=[':'])
data.replace(('M', 'F','I'), (0,1,2), inplace=True)
```

```python
data1 = data[feature_names].drop('Sex',axis=1)
```

```python
data2 = data1['Height'].values
data3 = data['Sex'].values
z1 = []
z2 = []
z3 = []
for i in range(len(data2)):
    if data3[i]==0:
        z1.append(data2[i])
    elif data3[i]==1:
        z2.append(data2[i])
    else:
        z3.append(data2[i])
plt.subplot(3,1,1)
plt.hist(z1)
plt.subplot(3,1,2)
plt.hist(z2)
plt.subplot(3,1,3)
plt.hist(z3)
plt.show()
```
