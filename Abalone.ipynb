```python
import pandas as pd
feature_names = ['Sex','Length','Diameter','Height','Whole Wt','Shucked Wt','Viscera Wt','Shell Wt','Rings']
```

```python
abalone = pd.read_csv('abalone.data',names = feature_names)
```

```python
abalone.head()
```

```python
abalone.describe()
```

```python
abalone.replace(('M', 'F','I'), (0,1,2), inplace=True)
```

```python
import keras
all_features = abalone[feature_names].drop('Sex',axis=1).values
all_classes = abalone['Sex'].values
all_classes = keras.utils.to_categorical(all_classes,3)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
all_features1 = scalar.fit_transform(all_features)
```

```python
from keras.optimizers import adam
from keras.layers import Dense,Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
import numpy as np
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(128,activation='relu',input_shape=(8,)))
    model.add(Dropout(0.4))
    model.add(Dense(256,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(256,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(128,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(32,activation='relu'))
    model.add(Dropout(0.4))
    model.add(Dense(3,activation='softmax'))
    model.compile(loss='categorical_crossentropy',
                  optimizer = adam(),
                  metrics = ['accuracy'])
    return model
```

```python
estimator = KerasClassifier(build_fn=create_model,nb_epoch=50,verbose=2)
score = cross_val_score(estimator,all_features1,all_classes,cv=50)
```
